%               You should set J to the cost.

````python
#Takes X with 2 columns and converts it into a polynomial with degree + 1
def mapFeature(x, degree=2):
    out = np.ones([1, x.shape[0]])
    for up in np.arange(1, degree+1):
        for down in np.arange(up+1):
            out = np.append(out, [x[:,0]**(up-down) * x[:,1]**(down)], axis=0)
    return out.T
                
````

````python
#Prepare data columns
m = np.size(data2['passed'])
X_mod = data2[['test1', 'test2']].values
Y_mod = data2['passed'].values
X_mod = mapFeature(X_mod, 6)
````

````python
#Cost function with regularization term
def computeCost(theta, X, y, l=1):
    '''Compute the cost function
    
    X = Data series containing x values
    y = Data series of results
    theta = linear regression parameters
    l = lambda or the regularization parameter
    
    returns cost
    '''
    tx = sigmoid(np.dot(X,theta))
    m = np.size(y)
    A = np.dot(-y, np.log(tx))
    B = np.dot((1-y), np.log(1 - tx))
    J0 = (A-B)/m
    J1 = (l/(2*m))*np.dot(theta.T[1:], theta[1:]) #dont reg. theta0
    J = J0+J1
    
    reg_theta = np.zeros_like(theta)
    reg_theta[1:] = (l/m)*theta[1:]
    grad = np.dot((tx-y),X)/m + reg_theta
    
    return J, grad ##some the minimizers expect an array
````

````python
#COmpute cost for the initial value
theta = np.zeros(X_mod.shape[1])
print "initial cost ", computeCost(theta, X_mod, Y_mod, 1)[0]
````

````python
#Minimize the cost function
func = lambda t: computeCost(t, X_mod, Y_mod, 1)
theta_init = np.zeros(X_mod.shape[1])
result = minimize(func, theta_init, method='Newton-CG', 
                  jac=True, options={'disp':False})
````

````python
#Optimal theta values
print result.x
````

````python
#Generate a meshgrid and calculate the cost on grid points
def generateMesh(theta):
    u = np.linspace(-1,1.5,50)
    v = np.linspace(-1,1.5,50)
    T0, T1 = np.meshgrid(u, v) #grid points

    J = []
    for (p,q) in zip(T0.ravel(), T1.ravel()):
        x = mapFeature(np.array([[p,q]]), degree=6)
        prob = np.dot(x, theta)
        #print p, q, cost
        J.append(prob[0])
    
    J = np.array(J).reshape(T0.shape) 
    return T0, T1, J
````

````python
#Do fitting for three extreme cases of lambda
theta_optim = []
for l in [1, 0.001, 100]:
    print "doing l: %d"%l
    func = lambda t: computeCost(t, X_mod, Y_mod, l)
    theta_init = np.zeros(X_mod.shape[1])
    result = minimize(func, theta_init, method='Newton-CG', 
                      jac=True, options={'disp':False})
    theta_optim.append(result.x)
    
````

````python
#Let the plotting fiesta begin
fig, ax = plt.subplots(ncols=3, figsize=(12,5))
for i,l in enumerate([1,0.001,100]):
    T0, T1, J = generateMesh(theta_optim[i])
    CS = ax[i].contour(T0, T1, J, [0], label='Decision Boundary')
    ax[i].plot(data2['test1'][passed], data2['test2'][passed],
                marker='x',lw=0, c='blue', label='Passed')
    ax[i].plot(data2['test1'][~passed], data2['test2'][~passed],
                marker='o',lw=0, c='red', label='Rejected')
    ax[i].set_xlabel('Test 1')
    ax[i].set_ylabel('Test 2')
    ax[i].set_xlim(-1.5,1.5)
    ax[i].set_ylim(-1.5,1.5)
    #ax.clabel(CS, inline=1, fontsize=10)
    ax[i].set_title(r'$\lambda=%.2f$'%l)
    ax[i].legend(numpoints=1, loc=3)

plt.tight_layout()
````
