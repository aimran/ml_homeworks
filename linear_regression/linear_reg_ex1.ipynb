###Compute Cost

``` matlab
function J = computeCost(X, y, theta)
%COMPUTECOST Compute cost for linear regression
%   J = COMPUTECOST(X, y, theta) computes the cost of using theta as the
%   parameter for linear regression to fit the data points in X and y

% Initialize some useful values
m = length(y); % number of training examples

% You need to return the following variables correctly
J = 0;

% ====================== YOUR CODE HERE ======================
% Instructions: Compute the cost of a particular choice of theta
%               You should set J to the cost.

% =========================================================================

end
```

````python
#Cost function
def computeCost(X, y, theta):
    '''Compute the cost function
    X = Data series containing x values
    y = Data series of results
    theta = linear regression parameters, size 2
    
    returns cost
    '''
    m = np.size(y)
    
    error = X.apply(lambda x: theta[0] + theta[1]*x)
    J = (error - y).apply(lambda x: np.square(x)).sum()
    return J/(2*m)
````

````python
# Check that for theta = (0,0), you get J(theta) = 32.07
X = data1['pop']
y = data1['profit']
print computeCost(X, y, np.array([0,0]))
print computeCost(X, y, np.array([-10,-1]))
````

###Gradient descent
``` matlab
function [theta, J_history] = gradientDescent(X, y, theta, alpha, num_iters)
%GRADIENTDESCENT Performs gradient descent to learn theta
%   theta = GRADIENTDESENT(X, y, theta, alpha, num_iters) updates theta by
%   taking num_iters gradient steps with learning rate alpha

% Initialize some useful values
m = length(y); % number of training examples
J_history = zeros(num_iters, 1);

for iter = 1:num_iters

    % ====================== YOUR CODE HERE ======================
    % Instructions: Perform a single gradient step on the parameter vector
    %               theta.
    %
    % Hint: While debugging, it can be useful to print out the values
    %       of the cost function (computeCost) and gradient here.
    %
    
    % ============================================================

    % Save the cost J in every iteration
    J_history(iter) = computeCost(X, y, theta);

end

end
```

````python
def gradientDescent(X, y, theta, alpha, num_iters):
    m = np.size(y)
    J_history = []
    
    for i in xrange(num_iters):
        error = X.apply(lambda x: theta[0] + theta[1]*x) - y
        theta[0] = theta[0] - alpha*error.sum()/m
        theta[1] = theta[1] - alpha*(error * X).sum()/m
        
        cost = computeCost(X,y,theta)
        J_history.append(cost)
    
    return theta, np.array(J_history)
````

````python
#Do a small test with 
theta0 = np.zeros(2)
t, h = gradientDescent(X,y,theta0, 0.01,5)
print t
print h
````

````python
#Do a full run
theta0 = np.zeros(2)
t,h = gradientDescent(X,y,theta0, 0.01, 1500)
````

````python
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(h)
ax.set_xlabel('# of iteration')
ax.set_ylabel(r'J($\theta$)')
plt.show()
````

````python
#Let the plotting fiesta begin!


linear_fit = lambda x,th: th[0] + th[1]*x

print 'best fit: theta0 %.2f  theta1: %.2f'%(t[0], t[1])


fig, ax = plt.subplots(figsize=(8,6))
ax.plot(data1['pop'], data1['profit'], marker='x', lw=0, c='red', label='Training data')
ax.plot(data1['pop'], linear_fit(data1['pop'], t), '-', c='#377EB8',lw=2, label='Linear fit')
ax.set_xlabel('Population of City in 10,000s')
ax.set_ylabel('Profit in \$10,000s')
ax.set_xlim(0,25)
ax.legend(loc=4, numpoints=1)
ax.grid()
plt.show()
````

###Visualizing J($\theta$)

````python
#Generate a meshgrid and calculate the cost on grid points
theta0_vals = np.linspace(-10,10,100)
theta1_vals = np.linspace(-1,4,100)
T0, T1 = np.meshgrid(theta0_vals, theta1_vals) #grid points

J = []
for (p,q) in zip(T0.ravel(), T1.ravel()):
    cost = computeCost(X,y,np.array([p,q]) )
    #print p, q, cost
    J.append(cost)
    
J = np.array(J).reshape(T0.shape) 
````

````python
fig, ax = plt.subplots(figsize=(6,6))
loc = [8,15,30,60,90]
CS = ax.contour(T0, T1, J, np.logspace(-2, 3, 20))
#ax.clabel(CS, inline=1, fontsize=10)
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.plot(t[0], t[1], 'rx', ms=10, lw=2)
ax.set_title(r'Plot of J($\theta$) as a function of $\theta_0$ and $\theta_1$')
plt.show()
````

````python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T0, T1,J, cmap=cm.hot)
ax.set_xlabel(r'$\theta_0$', labelpad=50)
ax.set_ylabel(r'$\theta_1$')
plt.show()
````
