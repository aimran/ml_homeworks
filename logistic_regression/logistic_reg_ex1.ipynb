````python
def sigmoid(x):
    return 1/(1+np.exp(-x))
````

``` matlab
function [J, grad] = costFunction(theta, X, y)
%COSTFUNCTION Compute cost and gradient for logistic regression
%   J = COSTFUNCTION(theta, X, y) computes the cost of using theta as the
%   parameter for logistic regression and the gradient of the cost
%   w.r.t. to the parameters.

% Initialize some useful values
m = length(y); % number of training examples

% You need to return the following variables correctly
J = 0;
grad = zeros(size(theta));

% ====================== YOUR CODE HERE ======================
% Instructions: Compute the cost of a particular choice of theta.
%               You should set J to the cost.
%               Compute the partial derivatives and set grad to the partial
%               derivatives of the cost w.r.t. each parameter in theta
%
% Note: grad should have the same dimensions as theta
%
```

````python
#Cost function
def computeCost(theta, X, y):
    '''Compute the cost function
    
    X = Data series containing x values
    y = Data series of results
    theta = linear regression parameters, size 2
    
    returns cost
    '''
    tx = sigmoid(np.dot(X,theta))
    m = np.size(y)
    A = np.dot(-y, np.log(tx))
    B = np.dot((1-y), np.log(1 - tx))
    J = (A-B)/m
    grad = np.dot((tx-y),X)/m
    return J, grad ##some the minimizers expect an array
````

````python
#add column of ones for the theta0 value
m = np.size(data1['accepted'])
X_mod = data1[['scoreA', 'scoreB']]
X_mod = np.hstack((np.ones((m,1)), X_mod.values))
Y_mod = data1['accepted'].values
theta = np.zeros(3)
````

````python
computeCost(theta, X_mod, Y_mod)
````

````python
fun1 = lambda t: computeCost(t, X_mod, Y_mod)[0]
fun2 = lambda t: computeCost(t, X_mod, Y_mod)

bfgc_min = lambda : minimize(fun2, np.zeros(3), method='BFGS', 
                             jac=True,  options={'disp':False})

nelder_mead_min = lambda : minimize(fun1, np.zeros(3), 
                                    method='Nelder-Mead', options={'disp':False})

powell_min = lambda : minimize(fun1, np.zeros(3), 
                               method='Powell', options={'disp':False})

newton_cg_min = lambda : minimize(fun2, np.zeros(3), 
                                  method='Newton-CG', jac=True, 
                                  options={'disp':False})
````
