````python
def sigmoid(z):
    return 1/(1+np.exp(-z))
````

````python
#Prepare data columns
m, n = data['X'].shape

X = np.hstack((np.ones((m,1)), data['X']))
print X.shape

initial_theta = np.zeros(n+1)
print initial_theta.shape

y = data['y'].flatten()
print y.shape
````

````python
print np.dot(X,initial_theta).shape
````

````python
#Cost function with regularization term
def computeCost(theta, X, y):
    '''Compute the cost function
    
    X = Data series containing x values
    y = Data series of results
    theta = linear regression parameters
    l = lambda or the regularization parameter
    
    returns cost
    '''
    m = np.size(y)
    
    #cost term
    tx = sigmoid(np.dot(X,theta)).flatten()
    J0 = np.dot(-y, np.log(tx))
    J1 = np.dot((1-y), np.log(1 - tx)) 
    cost = (J0-J1)
    
    #gradient
    grad = np.dot(X.T, (tx-y)) 
    return cost/m, grad/m 

def computeCost_reg(theta, X, y, ld=1):

    cost, grad = computeCost(theta, X, y)
    m = X.shape[0]
    reg_cost = (ld/(2*m))*np.dot(theta[1:].T, theta[1:]) #dont reg. theta0
    
    reg_grad = ld/m*theta
    reg_grad[0] = 0
    return cost+reg_cost, grad+reg_grad
    
````

````python
#Minimize the cost function for initial values
func = lambda t: computeCost_reg(t, X, 1*(y==2), 1)
result = minimize(func, initial_theta, 
                    method='CG', 
                    jac=True, 
                    options={
                        'maxiter':120,
                        'disp':True
                      }
                )

````

````python
def one_vs_all(X, y, num_labels, ld):
    n = X.shape[1]
    all_theta = np.zeros((num_labels, n))
    for label in range(1, num_labels+1):
        func = lambda t: computeCost_reg(t, X, 1*(y==label), ld)
        result = minimize(func, initial_theta, 
                            method='CG', 
                            jac=True, 
                            options={
                                'maxiter':120,
                                'disp':False,
                              }
                            )
        if not result.success:
            print "Training for label %d did not converge"%label
        else:
            print "Training for label %d -- cost: %.3f"%(label, result.fun)
        all_theta[label-1,:] = result.x
        
    return all_theta 
````

````python
all_theta = one_vs_all(X, y, 10, 1)
````

````python
def predictOneVsAll(all_theta, X):
    prob = sigmoid(np.dot(X, all_theta.T))
    prediction = 1 + np.argmax(prob, axis=1)
    return prediction

prediction =  predictOneVsAll(all_theta, X)
````

````python
accuracy = np.mean(prediction==y) * 100
print "Accuracy of the predictor is %.1f"%accuracy
````
